```python
import cv2
from ultralytics import YOLO
from IPython.display import display, Image, clear_output

# 1. تحميل الموديل
model = YOLO('correctionExercices.pt') # أو استخدم .pt للتجربة السريعة

# 2. فتح الكاميرا
cap = cv2.VideoCapture(0)

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # 3. تشغيل الموديل (تقليل الدقة لـ 320 لسرعة العرض في المتصفح)
        results = model(frame, imgsz=320, verbose=False)

        # 4. رسم النتائج
        annotated_frame = results[0].plot()

        # 5. تحويل الصورة لصيغة JPEG لعرضها في Jupyter
        _, buffer = cv2.imencode('.jpg', annotated_frame)

        # 6. تحديث العرض في الخلية
        clear_output(wait=True)
        display(Image(data=buffer))

except KeyboardInterrupt:
    # التوقف عند الضغط على زر التوقف في Jupyter
    print("تم إيقاف البث.")

finally:
    cap.release()
```
